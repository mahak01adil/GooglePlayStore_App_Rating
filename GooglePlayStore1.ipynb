# 4.1 Variables seem to have incorrect type and inconsistent formatting. You need to fix them: 

1. Size column has sizes in Kb as well as Mb. To analyze, you’ll need to convert these to numeric.

2. Extract the numeric value from the column

3. Multiply the value by 1,000, if size is mentioned in Mb

```python
# Printing the data of Size column to check MB and Kb
data['Size'].value_counts()
```

```python
# Removing Kb and MB from column Size
def change(Size):
    if 'M' in Size:
        x=Size[:-1]
        x=float(x)*1000
        return x
    elif 'k' in Size:
        x=Size[:-1]
        x=float(x)
        return x
    else:
        return None
        
```

```python
data['Size']=data['Size'].map(change)
```

```python
# Now checking MB and KB are removed from Size
data.head()
```

# 4.3 Installs field is currently stored as string and has values like 1,000,000+. 

1. Treat 1,000,000+ as 1,000,000

2. remove ‘+’, ‘,’ from the field, convert it to integer

```python
# Printing the data of Installs column
data['Installs']
```

```python
# Removing  ‘+’, ‘,’ from Installs column
data['Installs']=data['Installs'].str.replace('[+,]','',regex=True)
```

```python
# ‘+’, ‘,’ removed from Installs column
data.head()
```

```python
# Drop columns App, Last Updated, Current Ver, and Android Ver. These variables are not useful for our task
inp1.drop(["Last Updated","Current Ver","Android Ver","App"],inplace=True,axis=1)
```

```python
inp1.head()
```

```python
#Get dummy columns for Category, Genres, and Content Rating. This needs to be done as the models do not understand categorical data, and all data should be numeric. Dummy encoding is one way to convert character fields to numeric. Name of dataframe should be inp2
inp2=inp1.copy()
inp2
```

```python
inp2.dropna(inplace=True)
inp2.shape
```

```python
pd.get_dummies(inp2['Category'])
```

```python
pd.get_dummies(inp2['Content Rating'])
```

```python
pd.get_dummies(inp2['Genres'])
```

```python
#Train test split  and apply 70-30 split
X=inp2[['Reviews', 'Size', 'Installs', 'Price']]
y=inp2['Rating']
```

```python
X
```

```python
y
```

```python
# Separate the dataframes into X_train, y_train, X_test, and y_test.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
# linear regression 
from sklearn.linear_model import LinearRegression
reg = LinearRegression()  
reg.fit(X_train, y_train) 
```

```python
y_pred = reg.predict(X_test)
```

```python
dfReg = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
dfReg
```

```python
# Report the R2 on train set
from sklearn.metrics import mean_squared_error,r2_score
mse= mean_squared_error(y_test, y_pred)
print('mean_squared_error = ',mse)
r_score= r2_score(y_test,y_pred)
print('r2_score =',r_score)
```
